```python
import pandas as pd
import bibtexparser

with open('../data/external/records_extended.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)

extended_df = pd.DataFrame(bib_database.entries)
extended_df.dropna(subset="literature_review", inplace=True)

top_8 = ["European Journal of Information Systems", "Information Systems Journal", "Information Systems Research", "Journal of AIS", "Journal of Information Technology", "Journal of MIS", "Journal of Strategic Information Systems", "MIS Quarterly"]

df_1 = extended_df.loc[~extended_df['journal'].isin(top_8)]
df_1 = df_1[["title","year", "journal"]]

with open('../data/external/records.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)
    
records_df = pd.DataFrame(bib_database.entries)
records_df.dropna(subset="literature_review",inplace=True)

df_2 = records_df.loc[~records_df['journal'].isin(top_8)]
df_2 = df_2[["title","year"]]

papers = []


for index, row in df_1.loc[:,['title']].iterrows():
    #print(row.title)
    print(row)
    if row.title not in list(df_2['title']):
        papers.append(row.title)

print(len(papers))

df_1.to_csv('../data/excluded.csv')
```
